```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data for average temperatures of five cities over 10 years
cities = ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Miami']
avg_temps = [14.1, 18.3, 11.7, 20.2, 25.4]  # Average temperatures in Celsius

# Creating the bar chart
plt.figure(figsize=(8, 5))
plt.bar(cities, avg_temps, color=['blue', 'green', 'red', 'purple', 'orange'])
plt.xlabel('Cities')
plt.ylabel('Average Temperature (C)')
plt.title('Average Temperatures of Five Cities Over 10 Years')
plt.show()

# Data for study hours vs. student grades
np.random.seed(42)
study_hours = np.random.randint(1, 10, 50)  # Hours studied per week
grades = study_hours * 10 + np.random.randint(-5, 5, 50)  # Simulated grades

df = pd.DataFrame({'Study Hours': study_hours, 'Grades': grades})

# Creating the scatter plot
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Study Hours', y='Grades', data=df, color='blue', alpha=0.7)
plt.xlabel('Study Hours per Week')
plt.ylabel('Student Grades (%)')
plt.title('Correlation Between Study Hours and Student Grades')
plt.show()

# Analysis:
# - The bar chart shows that Miami has the highest average temperature while Chicago has the lowest.
# - The scatter plot indicates a positive correlation between study hours and student grades,
#   suggesting that students who study more tend to score higher.
```

```python
import yfinance as yf
import plotly.graph_objects as go
from datetime import datetime, timedelta
import time

# Define the ticker symbol and time frame
ticker_symbol = 'TSLA'
end_date = datetime.now()
start_date = end_date - timedelta(days=180)  # Last 6 months

# Function to fetch stock data with retries
def fetch_stock_data(ticker, start, end, retries=3, delay=5):
    for attempt in range(retries):
        try:
            stock_data = yf.download(ticker, start=start, end=end)
            if not stock_data.empty:
                return stock_data
        except Exception as e:
            print(f"Attempt {attempt + 1} failed: {e}")
        time.sleep(delay)
    return None

# Fetch stock data with retry logic
stock_data = fetch_stock_data(ticker_symbol, start_date, end_date)

# Ensure stock_data is not empty before plotting
if stock_data is not None and not stock_data.empty:
    # Create figure
    fig = go.Figure()

    # Add line trace for closing prices
    fig.add_trace(go.Scatter(
        x=stock_data.index,
        y=stock_data['Close'],
        mode='lines',
        name='Close Price',
        line=dict(color='blue'),
        hovertemplate='Date: %{x}<br>Close: $%{y:.2f}<extra></extra>'
    ))

    # Customize layout
    fig.update_layout(
        title=f'{ticker_symbol} Stock Prices Over the Last 6 Months',
        xaxis_title='Date',
        yaxis_title='Stock Price (USD)',
        hovermode='x',
        xaxis=dict(
            rangeselector=dict(
                buttons=list([
                    dict(count=1, label='1m', step='month', stepmode='backward'),
                    dict(count=3, label='3m', step='month', stepmode='backward'),
                    dict(step='all')
                ])
            ),
            rangeslider=dict(visible=True),
            type='date'
        )
    )

    # Show plot
    fig.show()
else:
    print("No stock data available for the given time frame. Check internet connection or try again later.")
```
